```python
%matplotlib inline
from datascience import *
import numpy as np
import matplotlib
import matplotlib.pyplot as plots

#plots.style.use('fivethirtyeight')
```

```python
pno = 1
total = 60

for i in np.arange(total): 
    pno= pno*(365-i)/365
```

```python
def birthday_prob(total):
   pno = 1
   if total >= 0 and total < 365:
     for i in np.arange(total): 
       pno= pno*(365-i)/365  # probability that no one shares a birthday
     return 1-pno # probability that at least two share a birthday
   elif total < 0:
     return 'must enter non-negative integer'
   else:
     return 1
```

```python
probs = []
for tot in np.arange(365):
    probs = np.append(probs,birthday_prob(tot+1))
```

```python
x = np.arange(365)+1;
plots.plot(x[0:99], probs[0:99])
plots.title('Probability of Shared Birthday')
plots.xlabel('Number of people in room')
plots.ylabel('Probability')
```
